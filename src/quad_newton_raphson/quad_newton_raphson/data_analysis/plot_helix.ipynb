```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# === Load data ===
# df = pd.read_csv('newhelix.log')
df = pd.read_csv('hx_spin.log')

# === Time bounds ===
start_time = df['time'].iloc[0] + 40.0
end_time = df['time'].iloc[-1] - 55.0
```

```python
# === Clip data to valid window ===
df_clipped = df[(df['time'] >= start_time) & (df['time'] <= end_time)]

# === Extract columns ===
x = df_clipped['x']
y = df_clipped['y']
z = df_clipped['z']
x_ref = df_clipped['x_ref']
y_ref = df_clipped['y_ref']
z_ref = df_clipped['z_ref']

# === Plot ===
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot3D(x_ref, y_ref, z_ref, 'b--', label='Reference')
ax.plot3D(x, y, z, 'r-', label='Actual')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Helix Trajectory (Clipped)')
ax.legend()
ax.set_box_aspect([1, 1, 1])
ax.invert_zaxis()

plt.show()
```

```python
z.max()# plot x v y ref and real and also z v y ref and real and x v z ref and real

fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs[0, 0].plot(x_ref, y_ref, 'b--', label='Reference')
axs[0, 0].plot(x, y, 'r-', label='Actual')
axs[0, 0].set_xlabel('X')
axs[0, 0].set_ylabel('Y')
axs[0, 0].set_title('X vs Y')
axs[0, 0].legend()


# plot z over time and zref over time
axs[0, 1].plot(df_clipped['time'], df_clipped['z'], 'r-', label='Z')
axs[0, 1].plot(df_clipped['time'], df_clipped['z_ref'], 'b--', label='Z Reference')
axs[0, 1].set_xlabel('Time')
axs[0, 1].set_ylabel('Z')
axs[0, 1].set_title('Z over Time')
axs[0, 1].legend()



axs[1, 0].plot(x_ref, z_ref, 'b--', label='Reference')
axs[1, 0].plot(x, z, 'r-', label='Actual')
axs[1, 0].set_xlabel('X')
axs[1, 0].set_ylabel('Z')
axs[1, 0].set_title('X vs Z')
axs[1, 0].legend()

#plot yaw and yawref over time
axs[1, 1].plot(df_clipped['time'], df_clipped['yaw'], 'g-', label='Yaw')
axs[1, 1].plot(df_clipped['time'], df_clipped['yaw_ref'], 'm--', label='Yaw Reference')
axs[1, 1].set_xlabel('Time')
axs[1, 1].set_ylabel('Yaw')
axs[1, 1].set_title('Yaw over Time')
axs[1, 1].legend()




plt.tight_layout()
plt.show()
# === Save the plots ===
# fig.savefig('helix_trajectory.png', dpi=300)
# fig.savefig('helix_trajectory_subplots.png', dpi=300)

```
